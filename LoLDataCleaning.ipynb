```python
import pandas as pd
from pathlib import Path
pd.set_option('display.max_columns', None)
```

```python
league_df = pd.read_csv(Path('2022_LoL_esports_match_data_from_OraclesElixir.csv'))
league_df.head(5) 
```

```python
# Checking Leagues
league_df['league'].unique()
```

```python
# Aquiring relevant columns for players
lcs_player_league_df = league_df[['gameid','league','year','split','game','participantid','side','position','playername','playerid','teamname','teamid','champion','result',
'kills','deaths','assists','dpm','damagetakenperminute','damagemitigatedperminute','wpm','vspm','earned gpm','cspm']]

# Filtering out team data
lcs_player_league_df = lcs_player_league_df[(lcs_player_league_df['position'] != 'team') & ((lcs_player_league_df['league'] == 'LCS') 
| (lcs_player_league_df['league'] == 'MSI') | (lcs_player_league_df['league'] == 'WLDs'))].reset_index(drop=True)

# Displaying data
lcs_player_league_df.tail(20)
```

```python
# Aquiring relevant columns for teams
lcs_team_league_df = league_df[['gameid','league','year','split','game','participantid','side','position','teamname','teamid','result',
'teamkills','teamdeaths','team kpm','firstblood','wpm','vspm','earned gpm','cspm']]

# Filtering out player data
lcs_team_league_df = lcs_team_league_df[(lcs_team_league_df['position'] == 'team') & ((lcs_team_league_df['league'] == 'LCS')
 | (lcs_team_league_df['league'] == 'MSI') | (lcs_team_league_df['league'] == 'WLDs'))].reset_index(drop=True)

# Displaying data
lcs_team_league_df.tail(20)
```
