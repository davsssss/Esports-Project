```python
import pandas as pd
```

```python
lock_in_df = pd.read_csv('b_2023_lock_in_brazil.csv')
masters_df = pd.read_csv('b_2023_masters_tokyo.csv')
champs_df = pd.read_csv('2023_champions.csv')
```

```python
champs_df.head()
```

```python
lock_in_df.columns
lock_in_df = lock_in_df.drop(columns = ['+/–', 'KAST', 'HS%', '+/–.1'])
lock_in_df.head()
```

```python
masters_df.columns
masters_df = masters_df.drop(columns = ['+/–', 'KAST', 'HS%', '+/–.1'])
masters_df.head()
```

```python
champs_df.columns
champs_df = champs_df.drop(columns = ['+/–', 'KAST', 'HS%', '+/–.1'])
champs_df.head()
```

```python
result_df = pd.concat([lock_in_df, masters_df, champs_df], ignore_index=True)
result_df.head(10)
```

```python
result_df['OriginalOrder'] = range(len(result_df))

# Group by 'Team', 'MapName', 'Result', and 'Margin'
grouped_df = result_df.groupby(['Team', 'MapName', 'Result', 'Margin'], as_index=False).agg({
    'OriginalOrder': 'first',  # Take the first value of the original order
    'R': 'mean',
    'ACS': 'mean',
    'K': 'mean',
    'D': 'mean',
    'A': 'mean',
    'ADR': 'mean',
    'FK': 'mean',
    'FD': 'mean',
    'VS': 'first'  # Since 'VS' is constant within a group, you can take the first value
})

# Sort by the original order column to restore the original order
grouped_df = grouped_df.sort_values(by='OriginalOrder').drop('OriginalOrder', axis=1).reset_index(drop=True)

# Display the result
print(grouped_df)
```

```python
grouped_df.to_csv('./valorant_match_predict/clean_formodel.csv', index=False)
```
