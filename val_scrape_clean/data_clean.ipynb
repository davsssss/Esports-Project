```python
desired_order = ['Team', 'Player', 'R', 'ACS', 'K', 'D', 'A', '+/–', 'KAST', 'ADR', 'HS%', 'FK', 'FD', '+/–.1' ]
df = df[desired_order]
df.columns
```

```python
df.dtypes
```

```python
# Convert 'KAST' to numeric by removing the percentage sign
#df['KAST'] = df['KAST'].str.extract(r'(\d+\.\d+|\d+)').astype(float)

# Columns with percentage signs
columns_with_percentage = ['ADR', 'HS%', 'FK', 'FD', '+/–.1']

# Loop through columns and convert to numeric by removing the percentage sign
for col in columns_with_percentage:
    df[col] = df[col].str.extract(r'(\d+\.\d+|\d+)').astype(float)

# Extract the first part of numeric values for the specified columns
columns_to_extract_first_number = ['R', 'ACS', 'K', 'D', 'A', '+/–']

for col in columns_to_extract_first_number:
    df[col] = df[col].astype(str).str.split().str[0].astype(float)
```

```python
df = df.dropna(subset=['R'])
```
